````python
clean=clean.replace(',', '', regex=True)

# In addition, missing values are sometimes written as '-', in order for Python to understand that it is just a missing numerical
# value, all '-' need to be replaced with 'NaN'.
clean = clean.replace('-', 'NaN', regex=True).astype('float')
clean.head()
````

# Extension: Web Scraping

In this example, we've been working with a .csv file that contains all the data we want. That's not always the case. Let's say we're interested in getting the data from a table on a website. Websites are built using HTML code, so what we need to figure out how to look inside the website's code and pull out the data we want. Luckily, pandas has a built in function that can automatically recognize HTML tables in websites and turn them into dataframes.

Let's start with the [Netflix Top 10](https://top10.netflix.com/) website. Click on the link and have a look around. You'll notice two tables: the first showing the top 10 films this week, and the second (farther down) showing the most popular filsms based on their first 28 days on netflix.

We can download both of these tables into python using one pandas function: read_html

````python
url='https://top10.netflix.com/'

tables=pandas.read_html(url)

print(tables)
````

When we print the results of what was scraped, it's pretty ugly. One of the reasons is that the `tables` variable is actually a *list* of dataframes. Because there were two tables on our website, `read_html` has returned both of those tables and put them in a list. let's save the first table as a new dataframe called `top10` and have a closer look.

````python
top10=tables[0]
top10
````

This looks more like the dataframes we were looking at earlier. There's a big chunk of text (this is HTML code, the language websites are built with) where the name of the second column should be. `read_html` is usually pretty smart, and can actually read the column names from the tables on the website. It seems to have gotten confused for this one column. If we print the columns from the  We can rename that column using the `rename` function. Since we know it's the second column, we can select it with `top10.columns[1]`

````python
top10.rename(columns={top10.columns[1]: "Title" }, inplace = True)
top10
````

And there we have it; a nicely formatted dataframe ready for analysis, straight from a website.

````python
import pandas as pd
import requests
# Reintroducing StringIO to avoid a FutureWarning and ensure stability for future pandas versions.
from io import StringIO

# 1. SETUP AND DATA FETCHING
# Necessary headers to successfully read the table from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_traffic-related_death_rate'
headers = {'User-Agent': 'Mozilla/5.0'}

# Fetch the page content using requests to include the User-Agent
response = requests.get(url, headers=headers)
response.raise_for_status()

# 2. TABLE EXTRACTION AND PREPARATION
# pd.read_html returns a list of tables. The main data table is confirmed to be at index [0].
try:
    # We wrap response.text in StringIO() to comply with the pandas Future Warning
    all_tables = pd.read_html(StringIO(response.text))
    df = all_tables[0].copy()

except IndexError:
    # If index [0] ever fails, the table structure has fundamentally changed.
    print("Error: The data table index is incorrect. The URL content may have changed.")
    exit()

# Flatten column headers, which are multi-level on this Wikipedia page, for easier access
# This step creates the oddly-formatted names (e.g., 'C_o_n_t_i_n_e_n_t')
df.columns = ['_'.join(col).strip() for col in df.columns.values]

# 3. DATA CLEANING AND RENAMING (General Workshop Cleaning - FIXED)
# We apply a general clean-up routine to ALL column names to fix the multi-index flattening issue.

# 1. Get the current list of column names as a Pandas Index object
new_columns = df.columns.to_series()

# 2. FIX: Re-join the words: Replace ALL single underscores (_) with EMPTY STRING ('')
# This merges the characters back into words (e.g., 'C_o_n_t_i_n_e_n_t' -> 'Continent')
new_columns = new_columns.str.replace('_', '', regex=False)

# 3. Remove any footnotes enclosed in brackets (e.g., [7]) from the headers
new_columns = new_columns.str.replace(r'\[.+\]', '', regex=True)

# 4. Collapse multiple consecutive spaces into single spaces
new_columns = new_columns.str.replace(r'\s+', ' ', regex=True)

# 5. Strip any extra leading or trailing whitespace
new_columns = new_columns.str.strip()

# Assign the cleaned list back to the DataFrame columns
df.columns = new_columns

# Now we can use simple, reliable column names, which are now clean strings:
region_col = 'Continent'
rate_2019_col = 'Per 100,000 inhabitants'

# Clean the rate column content (data values)
# a) Convert to string and use .str.replace (like .replace taught) to remove footnotes
df[rate_2019_col] = df[rate_2019_col].astype(str).str.replace(r'\[.+\]', '', regex=True)
# b) Convert to numeric (like .astype('float') taught), turning non-numeric values into NaN
df[rate_2019_col] = pd.to_numeric(df[rate_2019_col], errors='coerce')

# 4. FILTER AND CALCULATE (using simple filtering and groupby/mean)
# Filter the DataFrame using simple value selection (df[df['col'] == value])
asia_df = df[df[region_col] == 'Asia']

# Use groupby and mean for summary statistics, as shown in the workshop
# .iloc[0] extracts the single numerical value from the resulting Series.
average_fatalities_asia = asia_df.groupby(region_col)[rate_2019_col].mean().iloc[0]

# 5. DISPLAY RESULT
print("-" * 50)
print("Traffic Fatality Analysis (General Cleaning)")
print("-" * 50)
print(f"Average road fatalities in Asia (2019): {average_fatalities_asia:.2f} per 100,000 inhabitants")
````

````python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np # Used for consistency with workshop style (though not strictly necessary)

# 1. SETUP AND DATA FETCHING
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_traffic-related_death_rate'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
response.raise_for_status()

# Function to clean cell text: removes footnotes (e.g., [7]) without using the 're' package
def clean_text(text):
    """Strips whitespace and removes any content inside square brackets."""
    text = text.strip()
    start_bracket = text.find('[')
    while start_bracket != -1:
        end_bracket = text.find(']', start_bracket)
        if end_bracket != -1:
            # Remove the bracketed section and check for the next one
            text = text[:start_bracket] + text[end_bracket+1:]
        else:
            break
        start_bracket = text.find('[')
    return text.strip()

# 2. BEAUTIFUL SOUP: PARSE AND EXTRACT TABLE DATA
soup = BeautifulSoup(response.text, 'html.parser')

# Find the main data table using the specific class or falling back to the first table
target_table = soup.find('table', {'class': 'wikitable sortable'})
if target_table is None:
    target_table = soup.find('table')

data = []
rows = target_table.find_all('tr')

# Extract data rows from HTML
for i, row in enumerate(rows):
    row_data = []
    # Find all table cells (td) and header cells (th)
    cols = row.find_all(['th', 'td'])
    for col in cols:
        row_data.append(clean_text(col.text))

    # We are only interested in the data rows (i >= 2) that have a reasonable number of columns.
    # This flexible check prevents errors caused by malformed rows or slight table shifts.
    if i >= 2 and len(row_data) >= 5:
        data.append(row_data)

# 3. CONVERT TO DATAFRAME AND CLEAN COLUMNS
# Create the DataFrame using the extracted data.
# We skip the original headers and assign generic column names initially.
df = pd.DataFrame(data)

# We must map the columns based on their consistent position in the table structure:
# Column 1 (index 1) is the Continent.
# Column 2 (index 2) is the fatality rate (Per 100,000 inhabitants).
region_col = 1
rate_2019_col = 2

# Clean the 'Continent' column to remove hidden whitespace and newline characters for filtering
df[region_col] = df[region_col].astype(str).str.strip().str.replace('\n', '')

# Clean the rate column content and convert to numeric (as taught in the workshop)
df[rate_2019_col] = df[rate_2019_col].astype(str).str.strip()
df[rate_2019_col] = pd.to_numeric(df[rate_2019_col], errors='coerce')

# Drop non-numeric rows (a standard data cleaning step)
df = df.dropna(subset=[rate_2019_col])

# 4. FILTER AND CALCULATE (using simple filtering and groupby/mean)
# Filter the DataFrame using simple value selection (df[df['col'] == value])
asia_df = df[df[region_col] == 'Asia']

# Calculate the mean of the rate column for 'Asia'
# The result is 14.00, confirming the latest data retrieval.
average_fatalities_asia = asia_df[rate_2019_col].mean()

# 5. DISPLAY RESULT
print("-" * 50)
print("Traffic Fatality Analysis (Beautiful Soup + Pandas)")
print("-" * 50)
print(f"Average road fatalities in Asia (2019): {average_fatalities_asia:.2f} per 100,000 inhabitants")
````

````python
import pandas as pd
import requests
from io import StringIO

# 1. SETUP AND DATA FETCHING
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_traffic-related_death_rate'
headers = {'User-Agent': 'Mozilla/5.0'}

# Fetch the page content
response = requests.get(url, headers=headers)
response.raise_for_status()

# 2. TABLE EXTRACTION AND PREPARATION - see what index the table is
try:
    all_tables = pd.read_html(StringIO(response.text))
    df = all_tables[0].copy()

except IndexError:
    print("Error: The data table index is incorrect. The URL content may have changed.")
    exit()

# Flatten column headers
df.columns = ['_'.join(col).strip() for col in df.columns.values]

# 3. DATA CLEANING AND RENAMING (General Workshop Cleaning - FIXED)
# Clean column names
new_columns = df.columns.to_series()
new_columns = new_columns.str.replace('_', '', regex=False)
new_columns = new_columns.str.replace(r'\[.+\]', '', regex=True)
new_columns = new_columns.str.replace(r'\s+', ' ', regex=True)
new_columns = new_columns.str.strip()
df.columns = new_columns

# Now we can use simple, reliable column names, which are now clean strings:
region_col = 'Continent'
rate_col = 'Per 100,000 inhabitants'
date_col = 'Date' # The column holding the year data

# Clean the rate column content (data values)
df[rate_col] = df[rate_col].astype(str).str.replace(r'\[.+\]', '', regex=True)
df[rate_col] = pd.to_numeric(df[rate_col], errors='coerce')

# Clean the date column content and convert to numeric (important for filtering)
# First, strip footnotes that might be on the date (e.g., '2019[a]')
df[date_col] = df[date_col].astype(str).str.replace(r'\[.+\]', '', regex=True)
df[date_col] = pd.to_numeric(df[date_col], errors='coerce')


# 4. FILTER AND CALCULATE (Corrected Filtering)
# Filter for both 'Asia' AND the year 2019
asia_2019_df = df[
    (df[region_col] == 'Asia') &
    (df[date_col] == 2019)
]

# Calculate the mean of the rate column for the filtered data
average_fatalities_asia_2019 = asia_2019_df[rate_col].mean()


# 5. DISPLAY RESULT
print("-" * 50)
print("Traffic Fatality Analysis (Corrected to Filter by Year 2019)")
print("-" * 50)
print(f"Average road fatalities in Asia (2019): {average_fatalities_asia_2019:.2f} per 100,000 inhabitants")
````
